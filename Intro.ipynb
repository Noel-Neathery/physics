# Was sind eigentlich smartiS?

Hallo liebe Physiklernende,

ihr befindet euch im "Einführungssmarti" (oder kurz: Intro).
Der Begriff **smartiS** steht für smarte und interaktive Skripte. Die **smartiS**, die ihr hier findet, sollen euch beim Verständnis von physikalischen Gesetzen unterstützen.

Wie sind die **smartiS** aufgebaut?

- Jedes Smarti beginnt mit einer Überschrift, der Autorender, einer Aufzählung, sowie einem Verweis auf die Hochschule Hamm-Lippstadt.

- Daraufhin folgt die Überschrift des jeweiligen physikalischen Themas, das vom smarti behandelt wird, sowie das Inhaltsverzeichnis des smartis.

- smartiS beinhalten immer Erklärungstexte und Codezeilen, die das smarti interaktiv machen. Außerdem gibt es hier Bilder, Tabellen, Buttons und Diagramme.

- Was für euch besonders interessant sein dürfte: Um euer Wissen zu überprüfen gibt es so genannte **"Knowledge Checks"**. Das sind Aufgaben mit denen ihr euch selber überprüfen und auf Klausuren vorbereiten könnt.

- Am Ende der smartiS gibt es eine Danksagung inklusive Benennung der Lizenz unter der das smarti veröffentlicht ist.

Die smartiS wurde im Rahmen des Programms [Fellowships für Innovationen in der digitalen Hochschullehre](https://www.stifterverband.org/digital-lehrfellows-nrw) des Ministeriums für Kultur und Wissenschaft des Landes Nordrhein-Westfalen und des Stifterverbandes entwickelt. Weitere Details finden sich im [Projektantrag](https://www.stifterverband.org/digital-lehrfellows-nrw/2019/temmen_kersten).




# Wie setze ich die smartiS am besten ein?

Die von uns mithilfe der Software [Jupyter Notebook](https://jupyter.org/) erstellten smartiS könnt ihr am besten mit der Web-Applikation [nbviewer](https://nbviewer.jupyter.org/) aufrufen und ansehen.
![Jupyter](https://jupyter.org/assets/main-logo.svg) 

Im Eingabefeld "Enter the location of a Jupyter Notebook to have it rendered here:" gebt ihr dananach "PeterAKersten/physics" ein und bestätigt eure Eingabe durch Anklicken der **Go!** Taste.

Da unsere Skripte **interaktiv** sind, sollen euch hier individuelle Lösungen geboten werden. Um diese individuellen Lösungen zu ermöglichen müsst ihr die smartiS starten. Hierzu gibt es den Webservice [mybinder](https://mybinder.org/), den ihr über den nbviewer durch Anlicken des mybinder Logo auf der oberen Menüleiste starten könnt.

![mybinder](https://mybinder.org/static/logo.svg?v=f9f0d927b67cc9dc99d788c822ca21c0)

Wenn mybinder die Skripte geladen und verarbeitet hat (das kann beim ersten Laden ein paar Sekunden dauern), seht ihr das Directory und ihr könnt eins der smartiS durch Anklicken auswählen.

Jetzt noch das Symbol (>>) "Restart the kernel" anklicken und "Restart and Run all cells" bestätigen. Daraufhin solltet ihr keinen Code mehr sehen, sondern lediglich Erklärungstext, Bilder, Diagramme und Co.
Außerdem gibt es einen Button auf dem "Quellcode anzeigen" steht. Diesen Button braucht ihr nur, wenn ihr ausdrücklich dazu aufgefordert werden bzw. wenn euch interessiert, was hinter den interaktiven Teilen der Smartis steckt.

Nun könnt ihr nach Lust und Laune die verschiedenen Schieberegler und Buttons verwenden, um individuell und interaktiv eure Aufgaben zu lösen. Selbst den Quellcode könnt ihr verändern und den jeweiligen Abschnitt mit einem Klick auf die Run (>|) Taste erneut ausführen.
